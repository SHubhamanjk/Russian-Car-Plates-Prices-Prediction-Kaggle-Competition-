```python
# Parse region code
def parse_region_code(plate):
    for length in [3, 2]:
        code = plate[-length:]
        for region, codes in REGION_CODES.items():
            if code in codes:
                return region, code
    return None, None

# Extract government features
def get_government_features(letters, number, region_code):
    for (gov_letters, (num_from, num_to), gov_region), details in GOVERNMENT_CODES.items():
        if gov_letters == letters and gov_region == region_code and num_from <= number <= num_to:
            return details
    return [None, 0, 0, 0]

def is_special_number(number):
    return 1 if str(number).strip('0') in {'111', '222', ..., '999'} else 0

# df['is_special_number'] = df['number'].apply(is_special_number)

# Feature extraction
def extract_features(plate):
    region, region_code = parse_region_code(plate)
    prefix = re.sub(f'{region_code}$', '', plate)
    match = re.match(r'([A-Za-z]?)(\d{3})([A-Za-z]{2,3})', prefix)
    letters = match.group(1) + match.group(3) if match else ''
    number = int(match.group(2)) if match else 0
    gov_features = get_government_features(letters.upper(), number, region_code)
    return {
        'region': region,
        'letters': letters.upper(),
        'number': number,
        'is_special_number': is_special_number(number),
        'has_advantage': gov_features[2] if gov_features else 0,
        'is_forbidden': gov_features[1] if gov_features else 0
    }
```

```python
import numpy as np
def compute_sample_weight(row):
    weight = 1
    weight += np.log1p(row['has_advantage'] * (100 / 1.158))
    weight += np.log1p(row['is_special_number'] * (100 / 5.134))
    weight += np.log1p(row['is_lucky_number'] * (100 / 16.282))
    weight += np.log1p(row['is_round_number'] * (100 / 10.142))
    weight += np.log1p(row['is_palindrome'] * (100 / 48.775))
    return weight

df['sample_weights'] = df.apply(compute_sample_weight, axis=1)
```
